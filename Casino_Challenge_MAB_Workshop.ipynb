```python

def plot_cumulative(rewards, title="Cumulative Reward"):
    plt.figure()
    plt.plot(np.cumsum(rewards))
    plt.title(title)
    plt.xlabel("Step")
    plt.ylabel("Cumulative Reward")
    plt.show()
```


## üé≤ Environment: Stationary Bernoulli Bandits

Ten arms, each with a hidden probability of reward. You won't see the true means during play, but we print them here for **instructor debugging/analysis**. You may comment this out during the competition.


```python

def make_stationary_bandit(n_arms=10, seed=42):
    rng = np.random.default_rng(seed)
    true_means = rng.random(n_arms)  # in [0,1)
    return true_means

# Instructor may reveal (comment out in live competition to keep secret)
SEED_ENV = 42  # üîí Keep this fixed across all students for fairness (Round 1)
TRUE_MEANS = make_stationary_bandit(seed=SEED_ENV)
print("DEBUG ‚Äî True means (hidden in competition):", np.round(TRUE_MEANS, 3))
```


## ü§ñ Agent: Œµ‚ÄëGreedy (Fixed or Decaying)

- With probability **Œµ**, explore a random arm.
- Otherwise, exploit the best arm found so far (highest estimated value).
- Estimates updated via **incremental sample average**.


```python

def epsilon_greedy(true_means, steps=1000, epsilon=0.1, seed=None):
    rng = np.random.default_rng(seed)
    n_arms = len(true_means)
    Q = np.zeros(n_arms)     # value estimates
    N = np.zeros(n_arms)     # counts
    rewards = np.zeros(steps, dtype=float)
    actions = np.zeros(steps, dtype=int)
    for t in range(steps):
        if rng.random() < epsilon:
            a = rng.integers(0, n_arms)
        else:
            a = int(np.argmax(Q))
        r = 1.0 if rng.random() < true_means[a] else 0.0
        N[a] += 1
        Q[a] += (r - Q[a]) / N[a]  # incremental mean
        rewards[t] = r
        actions[t] = a
    return rewards, actions, Q, N

def epsilon_greedy_decaying(true_means, steps=1000, eps_start=0.5, eps_end=0.05, seed=None):
    rng = np.random.default_rng(seed)
    n_arms = len(true_means)
    Q = np.zeros(n_arms)
    N = np.zeros(n_arms)
    rewards = np.zeros(steps, dtype=float)
    actions = np.zeros(steps, dtype=int)
    for t in range(steps):
        # Linear decay
        epsilon = eps_end + (eps_start - eps_end) * max(0, (steps - 1 - t)) / max(1, steps - 1)
        if rng.random() < epsilon:
            a = rng.integers(0, n_arms)
        else:
            a = int(np.argmax(Q))
        r = 1.0 if rng.random() < true_means[a] else 0.0
        N[a] += 1
        Q[a] += (r - Q[a]) / N[a]
        rewards[t] = r
        actions[t] = a
    return rewards, actions, Q, N
```


# üîÑ Round 2 ‚Äî Non‚ÄëStationary Casino (Competition)

**Twist:** The slot machines drift over time. Fixed exploitation can fail; adaptive exploration helps.

Two common adaptations:
- Keep **Œµ** from decaying too low (retain exploration).
- Use a **constant step size Œ±** (exponential moving average) to weight recent rewards more.


```python

def nonstationary_means(n_arms=10, seed=2025):
    # Initialize random means
    rng = np.random.default_rng(seed)
    return rng.random(n_arms)

def step_drift(means, drift_scale=0.01, rng=None):
    if rng is None:
        rng = np.random.default_rng()
    means = means + rng.normal(0, drift_scale, size=means.shape)
    return np.clip(means, 0.0, 1.0)

def epsilon_greedy_constant_alpha(steps=2000, n_arms=10, eps=0.1, alpha=0.1, seed_env=7, seed_agent=None, drift_scale=0.01):
    # Non-stationary env with drifting means
    rng_env = np.random.default_rng(seed_env)
    rng_agent = np.random.default_rng(seed_agent)
    means = rng_env.random(n_arms)
    Q = np.zeros(n_arms)
    rewards = np.zeros(steps, dtype=float)
    actions = np.zeros(steps, dtype=int)
    for t in range(steps):
        # choose action
        if rng_agent.random() < eps:
            a = rng_agent.integers(0, n_arms)
        else:
            a = int(np.argmax(Q))
        # reward from current means
        r = 1.0 if rng_env.random() < means[a] else 0.0
        # constant step-size update (EMA)
        Q[a] = Q[a] + alpha * (r - Q[a])
        rewards[t] = r
        actions[t] = a
        # drift environment
        means = step_drift(means, drift_scale=drift_scale, rng=rng_env)
    return rewards, actions, Q
```

```python

# === YOUR SETTINGS (edit) ===
NAME_R2 = "Your Name"
STEPS_R2 = 3000
EPS_R2 = 0.1     # keep some exploration alive
ALPHA_R2 = 0.1   # constant step size for non-stationarity
SEED_ENV_R2 = 2025  # shared across class
SEED_AGENT_R2 = 999
DRIFT_SCALE = 0.01  # magnitude of mean drift per step

# === RUN ===
rewards_r2, actions_r2, Q_r2 = epsilon_greedy_constant_alpha(
    steps=STEPS_R2, n_arms=10, eps=EPS_R2, alpha=ALPHA_R2,
    seed_env=SEED_ENV_R2, seed_agent=SEED_AGENT_R2, drift_scale=DRIFT_SCALE
)

total_r2 = int(np.sum(rewards_r2))
print(f"""
[Round 2]
Player: {NAME_R2}
Strategy: epsilon_greedy + constant_alpha (eps={EPS_R2}, alpha={ALPHA_R2})
Steps: {STEPS_R2}
Total Reward: {total_r2}
""")
plot_cumulative(rewards_r2, title="Round 2 ‚Äî Non‚ÄëStationary Cumulative Reward")
```

# Multi-Armed Bandit ‚Äî Casino Challenge (Results Comparison)

I ran the provided Casino Challenge notebook twice:
- **Round 1 (Stationary)** ‚Üí saved as `submissions_round1.csv`
- **Round 2 (Non-Stationary)** ‚Üí saved as `submissions_round2.csv`

Now I am comparing the two runs to see how the strategies behaved in stable vs changing environments.


```python
import pandas as pd

# load both files generated by the workshop
r1 = pd.read_csv("submissions_round1.csv")
r2 = pd.read_csv("submissions_round2.csv")

display(r1.head())
display(r2.head())

r1["round"] = "Round 1 (Stationary)"
r2["round"] = "Round 2 (Non-Stationary)"
```

```python
r1["avg_reward_per_step"] = r1["total_reward"] / r1["steps"]
r2["avg_reward_per_step"] = r2["total_reward"] / r2["steps"]
```

```python
combined = pd.concat([r1, r2], ignore_index=True)
display(combined)
```

```python
summary = (
    combined
    .groupby("round")[["total_reward", "avg_reward_per_step"]]
    .mean()
    .reset_index()
)
display(summary)
```

```python
best_per_round = (
    combined
    .sort_values(["round", "avg_reward_per_step"], ascending=[True, False])
    .groupby("round")
    .head(3)    # top 3 per round
)
display(best_per_round)
```

```python
import matplotlib.pyplot as plt

# take only the best strategy from each round
tops = best_per_round.groupby("round").head(1)

plt.figure(figsize=(10,3))
plt.bar(
    tops["strategy"] + " (" + tops["round"].str.extract(r'\((.*)\)')[0] + ")",
    tops["avg_reward_per_step"]
)
plt.xticks(rotation=45, ha="right")
plt.ylabel("Avg reward per step")
plt.title("Best strategy in each round")
plt.tight_layout()
plt.show()
```

##  Talking Points ‚Äî Casino Challenge Comparison

1. **What I did**  
   I ran the Casino Challenge twice ‚Äî once in a *stationary* environment where rewards stayed constant, and once in a *non-stationary* setup where rewards slowly changed.  
   The goal was to see how the same Œµ-greedy approach behaves under both conditions.

2. **What I noticed**  
   In the stationary round, the agent quickly learned which arm was best and kept exploiting it, earning a higher average reward per step (about **0.93**).  
   In the non-stationary round, the agent‚Äôs average reward dropped a bit (to **0.85**) because it had to keep exploring as the casino‚Äôs payouts shifted.  
   However, it still managed a higher **total reward** since it ran for more steps and kept adapting.

3. **Why this happens**  
   When the environment doesn‚Äôt change, it‚Äôs smart to focus on exploitation ‚Äî what worked yesterday still works today.  
   But when the environment drifts, old knowledge becomes outdated, so the agent must learn continuously.  
   Using a constant learning rate (Œ± = 0.1) helps it stay responsive to those changes.

4. **What the plot tells us**  
   The first bar (Œµ-greedy in the stationary casino) is slightly taller, showing better per-step performance.  
   The second bar (Œµ-greedy with constant Œ±) is close behind, proving that even though it sacrifices a bit of reward, it‚Äôs better suited for changing conditions.

5. **What I learned**  
   - There isn‚Äôt one ‚Äúbest‚Äù bandit strategy ‚Äî it depends on whether the world stays the same or changes.  
   - **Œµ-greedy** is great for stable environments.  
   - **Œµ-greedy with constant Œ±** is better when things evolve over time.  
   - In real life (like ads, recommendations, or A/B testing), you‚Äôd want a balance ‚Äî keep exploring a little but also adapt to new trends.

6. **Final thought**  
   This experiment really showed the trade-off between **stability and adaptability**.  
   A model that never explores can get stuck, but one that keeps adapting stays ready for change ‚Äî just like decision-making in real life.
